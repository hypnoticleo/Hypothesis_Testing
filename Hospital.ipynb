### Problem Satement

A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.
   
  Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from scipy import stats
import seaborn as sns

pd.set_option('display.max_rows', None)
plt.style.use('seaborn-darkgrid')
```

### Initial EDA

```python
df = pd.read_csv('LabTAT.csv')
df
```

```python
plt.figure(figsize=(10,25), dpi = 100)
plt.subplot(4,1,1)
sns.distplot(df['Laboratory 1'])

plt.subplot(4,1,2)
sns.distplot(df['Laboratory 2'])

plt.subplot(4,1,3)
sns.distplot(df['Laboratory 3'])

plt.subplot(4,1,4)
sns.distplot(df['Laboratory 4'])

plt.show()
```

```python
df.describe()
```

We can only use One-Way Annova because we have to compare sample means, if we use 2 sample test there are 4C2 combinations, the error generated in each test will we compounded and the end result won't be reliable

### Using Scipy

```python
stats.f_oneway(df['Laboratory 1'], df['Laboratory 2'], df['Laboratory 3'], df['Laboratory 4'])
```

### It is evident that there is enough evidence that there is infact a significant difference between the groups
